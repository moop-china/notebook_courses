* 教学目的

    金融统计学习方法这一节，将介绍三个重要的金融模型，本教程将介绍的是ARCH、GARCH金融模型。可能部分人已经在课堂或者其他地方学习过，但是这里要作为重点介绍的是，如何用python来实现他们。同时对于没有接触过的人，提供了少许介绍，也帮助学过的人进行回忆。

    ARCH模型是获得2003年诺贝尔经济学奖的计量经济学成果之一。被认为是最集中反映了方差变化特点而被广泛应用于金融数据时间序列分析的模型。目前所有的波动率模型中,ARCH类模型无论从理论研究的深度还是从实证运用的广泛性来说都是独一无二的，广泛地用于验证金融理论中的规律描述以及金融市场的预测和决策。

    同时，你将学习用到前面分析基础的知识，初步地用起来，这对于开始用python解决实际问题而言只是开始，但是并不简单，所以我们在教程里会引导你，给你作出表率。
    
* 教学材料

    我们需要用到的数据是上证指数的k线数据，在1.2采集数据中有提到如何搜集
    
    我们将用到绘图包matplotlib和seaborn，数据分析基础包scipy、numpy和pandas，统计知识的包statsmodels和arch，以及数据源tushare
    
* 课程主体，按章节排序，用1.1和2.3.4这种层级安排内容，用markdown和程序交替的方式，展开程序逻辑，在有必要的地方进行程序注释
