<b>理论基础</b>

ARIMA(p,d,q) 称为差分自回归移动平均模型，根据原序列是否平稳以及回归中所含部分的不同，包括移动平均过程(MA)、自回归过程(AR)、自回归移动平均过程(ARMA)和自回归滑动平均混合过程(ARIMA)。

>移动平均过程（MA(q)）

>自回归过程（AR(p)）

>自回归移动平均过程（ARMA(p,q)）

>自回归积分滑动平均过程 （ARIMA(p,d,q)）

##  <b>数据清洗</b>
预处理是当数据不符合模型需求时，做的一系列处理，包括：缺失值插补、数据类型变换、特征提取等

1. 转换Date列的数据格式为datetime
2. 设置Date列为dataframe的索引
3. 对close做差分（非平稳）
4. 对close做log转换（非正态）
5. 删除多余的列
6. 训练集和测试集划分

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df['Close_diff'] = df['Close']-df.shift()['Close']
df['Close_diff_log'] = np.log1p(df['Close'])-np.log1p(df.shift()['Close'])

df = df[['Close', 'Close_diff', 'Close_diff_log']]
df = df.dropna()
print(df.head(10))
train = df['2015-01-01':'2017-01-01']
test = df['2017-01-01':]

print('\n', len(train), len(test))
```

## <b>建模</b>
<b>第一步：拟合ARIMA(p,d,q)：确定d、p、q参数</b>

1. 可视化和ADF检测close_diff(close差分)是否平稳

2. 可视化和AIC准则计算参数p、q。


>q – The lag value where the ACF chart crosses the upper confidence interval for the first time. 

>p - The lag value where the PACF chart crosses the upper confidence interval for the first time.


```python
#iplot([Scatter(x=train.index, y=train['Close_diff'])])
train['Close_diff'].plot()
plt.show()
#平稳性检测
from statsmodels.tsa.stattools import adfuller as ADF
print('close差分的平稳性检测p-value: {}'.format(ADF(train['Close_diff'].dropna())[1]))
#p<a,平稳的
```

```python
#iplot([Scatter(x=train.index, y=train['Close_diff'])])
#自相关 ACF
plt = sm.graphics.tsa.plot_acf(train['Close_diff'], lags=40)
#偏相关 PACF
plt = sm.graphics.tsa.plot_pacf(train['Close_diff'], lags=40)
```

```python
#AIC 计算
res = sm.tsa.arma_order_select_ic(train['Close_diff'], ic='aic', trend='nc')
res
```

差分后close平稳,d =1，p=3,q=0

<b> 第二步：检验模型是否合适</b>

模型残差是随机的。残差是没趋势（d=0,平稳）不自相关（p=0）和偏相关（q=0）

```python
from statsmodels.tsa.arima_model import ARIMA

arima_3_1_0 = ARIMA(train['Close'].as_matrix(), order=(3, 1, 0)).fit(dist=False)
arima_3_1_0.params
```

```python

plt = sm.graphics.tsa.plot_acf(arima_3_1_0.resid, lags=40)
plt.show()
plt = sm.graphics.tsa.plot_pacf(arima_3_1_0.resid, lags=40)
plt.show()
```

残差的ACF和PACF所有点都落在了阴影区域（置信区间内）

<b>第三步：进行预测</b>

 定义了一个函数完成预测并可视化展示：
 

```python
def show_predict(train, test, n_pre, p, d, q): #预测n_pre天的收盘价
    arima_init = ARIMA(train['Close'].as_matrix(), order=(p, d, q)).fit(dist=False)
    ts = train['Close'].as_matrix()
    predictions = np.empty((0), dtype=np.float32) #先建立了一个空array
    for i in tqdm(range(n_pre)):  #加了一个进度条
        arima = ARIMA(ts, order=(p, d, q)).fit(dist=False)
        predict = arima.forecast()[0]
        predictions = np.hstack([predictions, predict])
        ts = np.hstack([ts, predict])
    nans = np.zeros(len(train))
    nans[:] = np.nan
    orgs = pd.concat([train['Close'], test[:n_pre]['Close']])
    orgs = pd.DataFrame({
        'Date': orgs.index,
        'Original': orgs.as_matrix(),
        'Prediction': np.hstack([nans, predictions])
    })
    orgs = orgs.set_index('Date')
    orgs.plot(color=['blue', 'red'])
    plt.show()
    return arima_init
```

```python
arima_3_1_0 = show_predict(train, test, 100, 3, 1, 0)
```
